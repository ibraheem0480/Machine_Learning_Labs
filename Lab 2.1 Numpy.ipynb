#### How Random Seed Works (NumPy)

np.random.seed(x) resets the random number generator to a fixed starting point.

Using the same seed → produces the same random numbers every time.

Seed affects only the next random operations after it is called.

In Jupyter Notebook, each cell continues the same random sequence.

So, if you set the seed in one cell and generate numbers in another, the result will change.

To get identical output in different cells, set the seed again before each random operation.

```python
np.random.seed(24)
a = np.random.rand(4)
a
```

# 3 NumPy Statistical Functions

```python
import numpy as np

arr = np.array([4, 9, 2, 7, 5])
print("Array:", arr)

# Mean
print("Mean:", arr.mean())

# Maximum value
print("Max:", arr.max())

# Minimum value
print("Min:", arr.min())

# Index of maximum value
print("Argmax:", arr.argmax())

# Index of minimum value
print("Argmin:", arr.argmin())

# Standard deviation
print("Standard Deviation:", arr.std())

# Variance
print("Variance:", arr.var())

# Median
print("Median:", np.median(arr))

# Sum of all elements
print("Sum:", arr.sum())

# Cumulative sum
print("Cumulative Sum:", arr.cumsum())

# Sorting
print("Sorted Array:", np.sort(arr))
```
